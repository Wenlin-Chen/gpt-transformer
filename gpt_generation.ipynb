```python
import torch
from transformer_models import GPT
from gpt_utils import process_text

device = 'cuda' if torch.cuda.is_available() else 'cpu'
torch.manual_seed(1337)
```

```python
with open('./data/text.txt', 'r', encoding='utf-8') as f:
    text = f.read()

vocab_size, encode, decode = process_text(text)
```

```python
batch_size = 16
block_size = 64
embed_size = 96
n_heads = 4
n_layers = 3
dropout = 0.2
bias = False

gpt = GPT(vocab_size, n_layers, n_heads, embed_size, block_size, dropout, bias)
gpt = gpt.to(device)
gpt.load_state_dict(torch.load("./ckpt/gpt.pt"))
gpt.eval()
```

```python
gpt.eval()
context = torch.zeros((1, 1), dtype=torch.long, device=device)
print(decode(gpt.generate(context, max_new_tokens=5000)[0].tolist()))
```
